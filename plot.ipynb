```python
import sqlite3
import pandas as pd
import numpy as np
import scipy as sp
import plotly.express as px
```

```python
con = sqlite3.connect("sky.db")
```

```python
def plot_bubbles(q):
    df = pd.read_sql_query(q, con)
    fig = px.scatter(df, x="timestamp", y="rssi", hover_data=['distance'])
    fig.update_yaxes(autorange="reversed")
    fig.show()


def plot_pie(q):
    df = pd.read_sql_query(q, con)
    # df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries'  # Represent only large countries
    fig = px.pie(df, names='rssi', title='Percentage of RSSI')
    fig.show()


def plot_bar(q):
    data = pd.read_sql_query(q, con)
    fig = px.bar(data, x='ts', color='pings', hover_data=['rssi'])
    fig.show()
```

```python
plot_bubbles("SELECT * FROM LOG_LINES GROUP BY timestamp, distance")
```

```python
plot_pie("SELECT * FROM LOG_LINES GROUP BY timestamp, distance")
```

```python
plot_bar("""
select count(*) as pings, strftime ('%Y-%m-%d %H:00:00',timestamp) ts , rssi
from (
	SELECT *
	FROM LOG_LINES
	where rssi <-56 and rssi > -69
	GROUP BY timestamp, distance )
GROUP BY ts, rssi
order by ts, rssi
""")
```
